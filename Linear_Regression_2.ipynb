```python
#importing the libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#Preprocessing the dataset
dataset = pd.read_csv("Linear_Regression_2.csv")
X = dataset.iloc[:, 0].values.reshape(-1, 1)
y = dataset.iloc[:, 1].values
```

```python
#Splitting the dataset
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 1/3, random_state = 0)
```

```python
#Training the model
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

```python
#Predicting the y
y_pred = regressor.predict(X_test)
print(np.round(np.concatenate((y_pred.reshape(len(y_pred), 1), y_test.reshape(len(y_test), 1)),1 ), decimals=1))
```

```python
#Visualizing the Training set
plt.scatter(X_train, y_train, color = "red", label = "Data points")
plt.plot(X_train, regressor.predict(X_train), color = "blue", label = "Regression Line")
plt.title("X vs Y (Train Set)")
plt.xlabel("X")
plt.ylabel("y")
plt.legend()
plt.show()
```

```python
#Visualizing the Training set
plt.scatter(X_test, y_test, color = "red", label = "Data points")
plt.plot(X_test, y_pred, color = "blue", label = "Regression Line")
plt.title("X vs Y (Test set)")
plt.xlabel("X")
plt.ylabel("y")
plt.legend()
plt.show()
```
